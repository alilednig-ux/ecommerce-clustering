```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
print("1. Ma'lumotlar yuklanmoqda (Biroz vaqt olishi mumkin)...")
# UCI omboridan to'g'ridan-to'g'ri yuklab olish
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
df_raw = pd.read_excel(url)
print("Yuklandi! Jami qatorlar soni:", len(df_raw))
```

```python
# MA'LUMOTLARNI TOZALASH
print("2. Ma'lumotlar tozalanmoqda...")
# ID yo'qlarini va bekor qilingan tranzaksiyalarni olib tashlaymiz
df_clean = df_raw.dropna(subset=['CustomerID'])
df_clean = df_clean[df_clean['Quantity'] > 0]
```

```python
# Jami narxni hisoblaymiz (Soni * Donasi narxi)
df_clean['Total_Price'] = df_clean['Quantity'] * df_clean['UnitPrice']
```

```python
# RFM JADVALINI YARATISH (ID BO'YICHA)
print("3. Mijozlar ID si bo'yicha RFM hisoblanmoqda...")
# Datasetdagi oxirgi sanadan 1 kun keyingi sanani "bugun" deb olamiz
latest_date = df_clean['InvoiceDate'].max() + dt.timedelta(days=1)
```

```python
# Mijoz ID si bo'yicha guruhlaymiz
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days, # Recency (Oxirgi xariddan o'tgan kun)
    'InvoiceNo': 'nunique',                                # Frequency (Xaridlar soni)
    'Total_Price': 'sum'                                   # Monetary (Jami tushum)
})

rfm.rename(columns={'InvoiceDate': 'Recency',
                    'InvoiceNo': 'Frequency',
                    'Total_Price': 'Monetary'}, inplace=True)
```

```python
# MA'LUMOTLARNI MODELLAR UCHUN TAYYORLASH
# Real datada asimmetriya kuchli bo'ladi. Uni logarifm yordamida tekislaymiz
rfm_log = np.log1p(rfm)

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)
```

```python
# MODELLARNI O'QITISH

print(f"4. Modellar {len(rfm)} ta unikal xaridorlar ustida o'qitilmoqda (Kuting)...")
models = {
    "K-Means": KMeans(n_clusters=4, random_state=42, n_init=10),
    "Hierarchical": AgglomerativeClustering(n_clusters=4),
    "DBSCAN": DBSCAN(eps=0.5, min_samples=10),
    "GMM": GaussianMixture(n_components=4, random_state=42),
    "K-Medoids": KMedoids(n_clusters=4, random_state=42, init='k-medoids++')
}
```

```python
df_results = rfm.copy() # Natijalar asl sonlarda turishi kerak
baho_natijalari = {}
```

```python
for name, model in models.items():
    if name == "GMM":
        model.fit(rfm_scaled)
        labels = model.predict(rfm_scaled)
    else:
        labels = model.fit_predict(rfm_scaled)

    df_results[f'{name}_Klaster'] = labels
```

```python
# Silhouette bahosi (faqat DBSCAN hamma narsani 1 ta klaster qilib qo'ymagan bo'lsa)
if len(np.unique(labels)) > 1:
    score = silhouette_score(rfm_scaled, labels)
    baho_natijalari[name] = score
else:
    baho_natijalari[name] = 0.0

eng_yaxshi_model = max(baho_natijalari, key=baho_natijalari.get)
```

```python
# VIZUALIZATSIYA (PCA bilan)

print("5. Grafiklar chizilmoqda...")
pca = PCA(n_components=2)
rfm_pca = pca.fit_transform(rfm_scaled)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, name in enumerate(models.keys()):
    # Safely get the score, defaulting to NaN if the key doesn't exist
    score_value = baho_natijalari.get(name, float('nan'))

    sns.scatterplot(
        x=rfm_pca[:, 0], y=rfm_pca[:, 1],
        hue=df_results[f'{name}_Klaster'],
        palette='tab10', ax=axes[i], legend='full', s=30, alpha=0.6
    )
    # Display 'N/A' if the score is NaN, otherwise format it
    if np.isnan(score_value):
        axes[i].set_title(f"{name}\n(Aniqlik bahosi: N/A)", fontsize=12, fontweight='bold')
    else:
        axes[i].set_title(f"{name}\n(Aniqlik bahosi: {score_value:.2f})", fontsize=12, fontweight='bold')

fig.delaxes(axes[5])
plt.tight_layout()
plt.show()
```

```python
# BIZNES XULOSA VA JADVAL

print(f"\n ENG YAXSHI MODEL: {eng_yaxshi_model}")
print("-" * 60)
print(" ENG YAXSHI MODEL BO'YICHA MIJOZLAR SEGMENTLARI (O'rtacha ko'rsatkichlar):")
```

```python
# Asl (logarifmlanmagan) raqamlar orqali biznes tahlil
biznes_tahlil = rfm.groupby(df_results[f'{eng_yaxshi_model}_Klaster']).agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(1)

biznes_tahlil.columns = ['Oxirgi xarid (kun avval)', "O'rtacha xaridlar soni", "O'rtacha tushum ($)", 'Mijozlar soni']
biznes_tahlil.index.name = "Klaster Raqami"
display(biznes_tahlil)
```
